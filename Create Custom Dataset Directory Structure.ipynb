```python
import os 
import numpy as np
import cv2 as cv
```

```python
train_folder_name = "train image"
test_folder_name = "test image"
```

```python
current_dir = os.getcwd()
print(current_dir)
```

```python
train_folder_path = os.path.join(current_dir, "dataset", train_folder_name)
test_folder_path = os.path.join(current_dir, "dataset", test_folder_name)
print(train_folder_path)
```

```python
train_images = os.listdir(train_folder_path)
test_images = os.listdir(test_folder_path)
train_images
```

```python
for i, image_name in enumerate(train_images):
    train_images[i] = os.path.join(current_dir, "dataset", train_folder_name, image_name)
    
for i, image_name in enumerate(test_images):
    test_images[i] = os.path.join(current_dir, "dataset", test_folder_path, image_name)
    
train_images[:5]
```

```python
def createFolder(data_path, train_path, val_path):
    
    if os.path.exists(data_path) == False:
        os.mkdir(data_path)
        print("created custom_dataset folder")
        print("path => ", data_path)
        
    if os.path.exists(train_path) == False:
        os.mkdir(train_path)
        print("created train folder")
        print("path => ", train_path)
    
    if os.path.exists(val_path) == False:
        os.mkdir(val_path)
        print("created val folder")
        print("path => ", val_path)
```

```python
data_path = os.path.join(current_dir, "traffic sign dataset")
train_path = os.path.join(current_dir, "traffic sign dataset", "train")
val_path = os.path.join(current_dir, "traffic sign dataset", "val")
    
createFolder(data_path, train_path, val_path)
```

```python
def creataData(train_images, val_images, data_path, train_path, val_path): #input train and val folder path
    #train
    for image_path in train_images:
        img = cv.imread(image_path)
        
        imageName = os.path.basename(os.path.normpath(image_path))
        outputFolder_path = os.path.join(train_path, imageName[:3])
        if os.path.exists(outputFolder_path) == False:
            os.mkdir(outputFolder_path)
            
        cv.imwrite(outputFolder_path+"/"+imageName, img)    #write image
    #val
    for image_path in val_images:
        img = cv.imread(image_path)
        
        imageName = os.path.basename(os.path.normpath(image_path))
        outputFolder_path = os.path.join(val_path, imageName[:3])
        if os.path.exists(outputFolder_path) == False:
            os.mkdir(outputFolder_path)
            
        cv.imwrite(outputFolder_path+"/"+imageName, img)    #write image
        
    #create folder that does not have test image
    os.mkdir(val_path+"/009")
    os.mkdir(val_path+"/018")
    os.mkdir(val_path+"/019")
    os.mkdir(val_path+"/033")
    print("done")
```

```python
creataData(train_images, test_images, data_path, train_path, val_path)
```
